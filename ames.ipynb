```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('AmesHousing.csv')
```

```python
missing = data.isnull().sum()
missing = missing[missing > 0]
missing.sort_values(inplace=True)
missing.plot.bar()
```

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.distplot(data['SalePrice'], bins=20)
```

```python
data['SalePrice'].describe()
```

```python
numeric_features = data.select_dtypes(include=[np.number])
numeric_features.columns
```

```python
categorical_features = data.select_dtypes(include=[np.object])
categorical_features.columns
```

```python
correlation = numeric_features.corr()
correlation['SalePrice']
```

```python
sns.heatmap(correlation,square=True, vmax=0.8)
```

```python
data["Garage Cars"].value_counts()
```

```python
data[["Garage Cars","SalePrice"]].corr()
```

```python
k = 11
cols = correlation.nlargest(k, 'SalePrice')['SalePrice'].index
cm = data[cols].corr()
f, ax = plt.subplots(figsize = (14,12))
sns.heatmap(cm, vmax = 0.8, linewidths=0.01, square=True, annot=True, cmap='viridis',
            linecolor="white", xticklabels=cols.values, annot_kws={'size':12}, yticklabels=cols.values)
```

```python
sns.scatterplot(x='Garage Cars', y='SalePrice', data=data)
```

```python
sns.regplot(x='Garage Cars', y='Garage Area', data=data, scatter=True, fit_reg=True)
```

```python
fig, ((ax1, ax2),(ax3, ax4),(ax5, ax6)) = plt.subplots(nrows=3, ncols=2, figsize=(14,10))
sns.regplot(x='Overall Qual', y='SalePrice', data=data, scatter=True, fit_reg=True, ax=ax1)
sns.regplot(x='Gr Liv Area', y='SalePrice', data=data, scatter=True, fit_reg=True, ax=ax2)
sns.regplot(x='Garage Area', y='SalePrice', data=data, scatter=True, fit_reg=True, ax=ax3)
sns.regplot(x='Full Bath', y='SalePrice', data=data, scatter=True, fit_reg=True, ax=ax4)
sns.regplot(x='Year Built', y='SalePrice', data=data, scatter=True, fit_reg=True, ax=ax5)
sns.regplot(x='Wood Deck SF', y='SalePrice', data=data, scatter=True, fit_reg=True, ax=ax6)
```
