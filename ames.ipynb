```python
# MODELO DE REGRESSÃƒO: ELASTICNET
```

```python
df = pd.read_csv('AMES_Final_DF.csv', index_col=[0])
```

```python
df.head()
```

```python
df.info()
```

```python
X = df.drop('SalePrice', axis=1)
```

```python
y = df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)

#random_state Ã© um parÃ¢metro que Ã© comumente encontrado em vÃ¡rias bibliotecas e funÃ§Ãµes relacionadas a geraÃ§Ã£o de nÃºmeros aleatÃ³rios em linguagens de programaÃ§Ã£o, 
# especialmente no contexto de aprendizado de mÃ¡quina. 
# Esse parÃ¢metro Ã© usado para controlar a semente (seed) que inicializa o gerador de nÃºmeros aleatÃ³rios.
```

```python
scaler = StandardScaler()

#StandardScaler Ã© uma classe da biblioteca scikit-learn em Python, que Ã© amplamente usada para prÃ©-processamento de dados em tarefas de aprendizado de mÃ¡quina. 
#Ela faz parte do mÃ³dulo sklearn.preprocessing e Ã© utilizada para realizar a padronizaÃ§Ã£o (normalizaÃ§Ã£o z-score) dos dados.
```

```python
scaled_x_train = scaler.fit_transform(X_train)

# O mÃ©todo fit_transform() combina os passos de ajuste e transformaÃ§Ã£o em uma Ãºnica chamada de funÃ§Ã£o. Ele primeiro ajusta o scaler aos dados de treinamento usando fit(), 
# e em seguida, aplica a transformaÃ§Ã£o aos prÃ³prios dados de treinamento usando transform(). O resultado Ã© a matriz de dados padronizados.
scaled_x_test = scaler.transform(X_test)

#  O mÃ©todo transform() Ã© aplicado aos dados de treinamento (ou quaisquer outros dados que vocÃª deseja padronizar). 
# Ele usa as estatÃ­sticas calculadas durante o ajuste (fit) para padronizar os dados. A transformaÃ§Ã£o calcula a diferenÃ§a entre cada valor e a mÃ©dia, e divide pelo desvio padrÃ£o. 
# Isso resulta em dados padronizados com mÃ©dia zero e desvio padrÃ£o de um.
```

```python
base_elastic_model = ElasticNet(max_iter=1000000)

# ElasticNet (RegressÃ£o)
# max_iter : Ã© a quantidade de iteraÃ§Ãµes (passos de otimizaÃ§Ã£o) que o algoritmo usarÃ¡ para convergir para os coeficientes ideais do modelo. 
# Especificamente, ele controla quantas vezes o algoritmo iterarÃ¡ atravÃ©s do processo de otimizaÃ§Ã£o para ajustar os coeficientes do modelo de regressÃ£o ElasticNet.
```

```python
# agora vou criar um objeto GridSearchCV
```

```python
param_grid={'alpha':[100], 'l1_ratio':[1]}

# 'alpha': Isso se refere ao hiperparÃ¢metro alpha do modelo ElasticNet. O ElasticNet Ã© uma tÃ©cnica de regressÃ£o que combina regularizaÃ§Ã£o L1 (LASSO) e L2 (Ridge). 
# O alpha controla o quanto da regularizaÃ§Ã£o Ã© aplicado ao modelo. Valores maiores de alpha implicam em mais regularizaÃ§Ã£o.

# 'l1_ratio': Isso se refere ao hiperparÃ¢metro l1_ratio do modelo ElasticNet. O l1_ratio controla a mistura entre a penalizaÃ§Ã£o L1 (LASSO) e a penalizaÃ§Ã£o L2 (Ridge). 
# Um valor de l1_ratio de 1 indica que apenas a penalizaÃ§Ã£o L1 estÃ¡ sendo usada, enquanto um valor de 0 indica que apenas a penalizaÃ§Ã£o L2 estÃ¡ sendo usada.
```

```python
grid_model = GridSearchCV(base_elastic_model, 
                          param_grid=param_grid, 
                          scoring='neg_mean_squared_error', 
                          cv=5, verbose=1)

# scoring='neg_mean_squared_error': Isso especifica a mÃ©trica que o GridSearchCV usarÃ¡ para avaliar o desempenho dos modelos testados. 
# No caso, estÃ¡ usando o erro quadrÃ¡tico mÃ©dio negativo (neg_mean_squared_error). 
# Como o GridSearchCV procura maximizar a mÃ©trica de avaliaÃ§Ã£o, usar o negativo do erro quadrÃ¡tico mÃ©dio significa que ele buscarÃ¡ minimizar o erro quadrÃ¡tico mÃ©dio real.

# cv=5: Isso define o nÃºmero de "dobras" ou "folds" na validaÃ§Ã£o cruzada. O nÃºmero 5 indica que o conjunto de dados serÃ¡ dividido em 5 partes, e o GridSearchCV irÃ¡ executar 5 iteraÃ§Ãµes, 
# cada vez usando 4 partes como treinamento e 1 parte como teste. Isso ajuda a avaliar o desempenho do modelo de forma robusta.

# verbose=1: Isso controla o nÃ­vel de detalhes que o GridSearchCV irÃ¡ imprimir durante a execuÃ§Ã£o. O valor 1 significa que ele imprimirÃ¡ mensagens para mostrar o progresso da busca em grade.  
```

```python
grid_model.fit(scaled_x_train, y_train)
```

```python
grid_model.best_params_

# grid_model.best_params_ Ã© um atributo fornecido pelo objeto GridSearchCV da biblioteca scikit-learn apÃ³s a conclusÃ£o de uma busca em grade. 
# Esse atributo contÃ©m um dicionÃ¡rio com os melhores valores de hiperparÃ¢metros encontrados durante a busca em grade, 
# ou seja, as combinaÃ§Ãµes de hiperparÃ¢metros que resultaram no melhor desempenho de acordo com a mÃ©trica de avaliaÃ§Ã£o especificada.

```

```python
y_prediction = grid_model.predict(scaled_x_test)

# predict(): Este Ã© um mÃ©todo de muitos modelos de regressÃ£o do scikit-learn, incluindo o ElasticNet. 
# Ele Ã© usado para fazer previsÃµes com base em novos dados de entrada. No caso da regressÃ£o, ele retorna os valores previstos do alvo (no caso, variÃ¡vel y).
```

```python
mean_absolute_error(y_test, y_prediction)
```

```python
np.sqrt(mean_squared_error(y_test, y_prediction))
```
