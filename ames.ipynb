```python
# MODELO DE REGRESSÃO: ELASTICNET
```

```python
df = pd.read_csv('AMES_Final_DF.csv', index_col=[0])
```

```python
df.head()
```

```python
df.info()
```

```python
X = df.drop('SalePrice', axis=1)
```

```python
y = df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)

#random_state é um parâmetro que é comumente encontrado em várias bibliotecas e funções relacionadas a geração de números aleatórios em linguagens de programação, 
# especialmente no contexto de aprendizado de máquina. 
# Esse parâmetro é usado para controlar a semente (seed) que inicializa o gerador de números aleatórios.
```

```python
scaler = StandardScaler()

#StandardScaler é uma classe da biblioteca scikit-learn em Python, que é amplamente usada para pré-processamento de dados em tarefas de aprendizado de máquina. 
#Ela faz parte do módulo sklearn.preprocessing e é utilizada para realizar a padronização (normalização z-score) dos dados.
```

```python
scaled_x_train = scaler.fit_transform(X_train)

# O método fit_transform() combina os passos de ajuste e transformação em uma única chamada de função. Ele primeiro ajusta o scaler aos dados de treinamento usando fit(), 
# e em seguida, aplica a transformação aos próprios dados de treinamento usando transform(). O resultado é a matriz de dados padronizados.
scaled_x_test = scaler.transform(X_test)

#  O método transform() é aplicado aos dados de treinamento (ou quaisquer outros dados que você deseja padronizar). 
# Ele usa as estatísticas calculadas durante o ajuste (fit) para padronizar os dados. A transformação calcula a diferença entre cada valor e a média, e divide pelo desvio padrão. 
# Isso resulta em dados padronizados com média zero e desvio padrão de um.
```

```python
base_elastic_model = ElasticNet(max_iter=1000000)

# ElasticNet (Regressão)
# max_iter : é a quantidade de iterações (passos de otimização) que o algoritmo usará para convergir para os coeficientes ideais do modelo. 
# Especificamente, ele controla quantas vezes o algoritmo iterará através do processo de otimização para ajustar os coeficientes do modelo de regressão ElasticNet.
```

```python
# agora vou criar um objeto GridSearchCV
```

```python
param_grid={'alpha':[100], 'l1_ratio':[1]}

# 'alpha': Isso se refere ao hiperparâmetro alpha do modelo ElasticNet. O ElasticNet é uma técnica de regressão que combina regularização L1 (LASSO) e L2 (Ridge). 
# O alpha controla o quanto da regularização é aplicado ao modelo. Valores maiores de alpha implicam em mais regularização.

# 'l1_ratio': Isso se refere ao hiperparâmetro l1_ratio do modelo ElasticNet. O l1_ratio controla a mistura entre a penalização L1 (LASSO) e a penalização L2 (Ridge). 
# Um valor de l1_ratio de 1 indica que apenas a penalização L1 está sendo usada, enquanto um valor de 0 indica que apenas a penalização L2 está sendo usada.
```

```python
grid_model = GridSearchCV(base_elastic_model, 
                          param_grid=param_grid, 
                          scoring='neg_mean_squared_error', 
                          cv=5, verbose=1)

# scoring='neg_mean_squared_error': Isso especifica a métrica que o GridSearchCV usará para avaliar o desempenho dos modelos testados. 
# No caso, está usando o erro quadrático médio negativo (neg_mean_squared_error). 
# Como o GridSearchCV procura maximizar a métrica de avaliação, usar o negativo do erro quadrático médio significa que ele buscará minimizar o erro quadrático médio real.

# cv=5: Isso define o número de "dobras" ou "folds" na validação cruzada. O número 5 indica que o conjunto de dados será dividido em 5 partes, e o GridSearchCV irá executar 5 iterações, 
# cada vez usando 4 partes como treinamento e 1 parte como teste. Isso ajuda a avaliar o desempenho do modelo de forma robusta.

# verbose=1: Isso controla o nível de detalhes que o GridSearchCV irá imprimir durante a execução. O valor 1 significa que ele imprimirá mensagens para mostrar o progresso da busca em grade.  
```

```python
grid_model.fit(scaled_x_train, y_train)
```

```python
grid_model.best_params_

# grid_model.best_params_ é um atributo fornecido pelo objeto GridSearchCV da biblioteca scikit-learn após a conclusão de uma busca em grade. 
# Esse atributo contém um dicionário com os melhores valores de hiperparâmetros encontrados durante a busca em grade, 
# ou seja, as combinações de hiperparâmetros que resultaram no melhor desempenho de acordo com a métrica de avaliação especificada.

```

```python
y_prediction = grid_model.predict(scaled_x_test)

# predict(): Este é um método de muitos modelos de regressão do scikit-learn, incluindo o ElasticNet. 
# Ele é usado para fazer previsões com base em novos dados de entrada. No caso da regressão, ele retorna os valores previstos do alvo (no caso, variável y).
```

```python
mean_absolute_error(y_test, y_prediction)
```

```python
np.sqrt(mean_squared_error(y_test, y_prediction))
```
